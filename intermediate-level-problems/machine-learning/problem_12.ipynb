# **Problem Statement**  
## **12. Build a basic SVM classifier using sklearn and visualize the hyperplane.**

Build a basic Support Vector Machine (SVM) classifier using sklearn and visualize the decision boundary (hyperplane) for a 2D dataset.

### Constraints & Example Inputs/Outputs

### Constraints
- Use sklearn SVM
- Dataset must be 2D for visualization
- Binary classification
- Kernel: Linear

### Example Input:
```python
X = [[2, 3], [3, 4], [4, 5], [6, 5], [7, 7]]
y = [0, 0, 0, 1, 1]

```

### Expected Output:
- Trained SVM model
- Decision boundary (hyperplane)
- Support vectors highlighted
- Correct class separation

### Solution Approach

### What is SVM?
SVM finds a hyperplane that:
- Maximizes margin between classes
- Uses support vectors (closest points)
- Works well in high-dimensional spaces

### Steps to Solve
1. Generate or load dataset
2. Train linear SVM
3. Extract model parameters
4. Plot:
    - Data points
    - Decision boundary
    - Margin lines
    - Support vectors

### Solution Code

#### Approach1: Conceptual / Brute Force (Explanation Only)
SVM optimization involves quadratic programming and is not implemented manually here due to complexity.

#### Conceptual brute-force idea:
- Try multiple separating lines
- Measure margin
- Pick the one with maximum margin

➡️ Used only for theoretical understanding

### Alternative Solution

````python
# Approach2: Optimized (sklearn Linear SVM)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

def train_svm(X, y):
    model = SVC(kernel='linear', C=1.0)
    model.fit(X, y)
    return model
````

````python
# Hyperplane Visualization

def plot_svm_decision_boundary(model, X, y):
    plt.figure(figsize=(8, 6))

    # Scatter points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=50)

    # Plot support vectors
    plt.scatter(model.support_vectors_[:, 0],
                model.support_vectors_[:, 1],
                s=150, facecolors='none', edgecolors='k')

    # Create grid
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = model.decision_function(xy).reshape(XX.shape)

    # Plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k',
               levels=[-1, 0, 1], alpha=0.7,
               linestyles=['--', '-', '--'])

    plt.title("SVM Decision Boundary with Margins")
    plt.show()
````

### Alternative Approaches

### Test Case

````python
# Test Case 1: Simple Linearly Separable Data

X = np.array([
    [2, 3],
    [3, 4],
    [4, 5],
    [6, 5],
    [7, 7]
])
y = np.array([0, 0, 0, 1, 1])

model = train_svm(X, y)
plot_svm_decision_boundary(model, X, y)
````

````python
# Test Case 2: Synthetic Dataset

from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=1.2)

model = train_svm(X, y)
plot_svm_decision_boundary(model, X, y)

````

````python
# Test Case 3: Predict New Points

new_points = np.array([[3, 3], [6, 6]])
predictions = model.predict(new_points)

print("Predictions:", predictions)
````

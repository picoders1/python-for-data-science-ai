# **Problem Statement**  
## **3. Simulate a probability distribution and visualize it using Matplotlib.**

### Problem Statement

Simulate a probability distribution using random sampling and visualize the distribution using Matplotlib.

The goal is to understand how theoretical probability distributions emerge from random samples.

### Constraints & Example Inputs/Outputs

### Constraints
- Use random sampling
- Dataset size ≥ 100
- Visualization using Matplotlib only
- Distribution must be visible and interpretable

### Example:
- Distribution: Normal (Gaussian)
- Mean = 0
- Standard Deviation = 1
- Number of samples = 10,000

### Expected Output:
- Histogram approximating a bell curve
- Mean close to 0
- Standard deviation close to 1

### Solution Approach

**Step 1: Choose a distribution**

Common choices:
- Uniform
- Normal (Gaussian)
- Binomial
- Exponential

We’ll start with Normal Distribution.

**Step 2: Generate random samples**
- Brute force → use Python random
- Optimized → use NumPy vectorization

**Step 3: Visualize**
- Plot histogram
- Normalize histogram to represent probability density
- Overlay theoretical curve (optional)


### Solution Code

````python
# Approach1: Brute Force Code

import random
import matplotlib.pyplot as plt

def simulate_normal_distribution_bruteforce(n, mean=0, std=1):
    samples = []
    for _ in range(n):
        samples.append(random.gauss(mean, std))
    return samples
````

### Alternative Solution

````python
# Approach2: Optimized Simulation (Using Numpy)

import numpy as np
import matplotlib.pyplot as plt

def simulate_normal_distribution_optimized(n, mean=0, std=1):
    return np.random.normal(mean, std, n)
````

### Alternative Approaches

```python
| Distribution | NumPy Function            |
| ------------ | ------------------------- |
| Uniform      | `np.random.uniform()`     |
| Normal       | `np.random.normal()`      |
| Binomial     | `np.random.binomial()`    |
| Exponential  | `np.random.exponential()` |
```

### Test Case

````python
# Test Case1: Brute Force Simulation
samples = simulate_normal_distribution_bruteforce(
    n=10000, mean=0, std=1
)

plt.hist(samples, bins=50, density=True)
plt.title("Normal Distribution (Brute Force)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
````

````python
# Test Case2: Optimized Simulation
samples = simulate_normal_distribution_optimized(
    n=10000, mean=0, std=1
)

plt.hist(samples, bins=50, density=True)
plt.title("Normal Distribution (Optimized - NumPy)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
````

````python
# Test Case 3: Compare Mean & Std
print("Mean:", np.mean(samples))
print("Std Dev:", np.std(samples))
````

````python
# Test Case 4: UNiform Distribution
samples = np.random.uniform(low=0, high=1, size=10000)

plt.hist(samples, bins=50, density=True)
plt.title("Uniform Distribution")
plt.show()
````

````python
# Test Case 5: Exponential Distribution

samples = np.random.exponential(scale=1.0, size=10000)

plt.hist(samples, bins=50, density=True)
plt.title("Exponential Distribution")
plt.show()
````

### Expected Outputs
- Histogram visually matches expected distribution
- Mean and variance converge as sample size increases
- Optimized approach is significantly faster

## Complexity Analysis

Let n=number of samples

### Time Complexity
- Brute force: O(n)
- Optimized (NumPy): O(n) (faster due to vectorization)

### Space Complexity

O(n) – storing samples

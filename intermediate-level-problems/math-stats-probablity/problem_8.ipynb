# **Problem Statement**  
## **8. Generate synthetic data following a normal distribution.**

### Problem Statement

Generate synthetic data that follows a normal (Gaussian) distribution using Python, and validate that the generated data approximately matches the expected statistical properties such as mean, standard deviation, and shape of the distribution.

### Constraints & Example Inputs/Outputs

### Constraints
- Mean (μ) can be any real number
- Standard deviation (σ) must be positive
- Number of samples n ≥ 1
- Randomness should be reproducible using a random seed- Non-linear classifiers may produce complex boundaries

### Example Inputs:
```python
Mean (μ) = 50
Standard Deviation (σ) = 10
Number of samples (n) = 10,000
```

### Expected Output:
- Generated data has:
    - Mean ≈ 50
    - Std ≈ 10
- Histogram resembles a bell curve
- Distribution symmetric around the mean

### Solution Approach

**Step1: Understand Normal Distribution**
A normal distribution is defined by:
- Mean (μ)
- Standard deviation (σ)

```python
X∼N(μ,σ^2)
```

**Step2: Decide Generation Strategy**

Two approaches:
- Brute Force: Use the Box–Muller transform
- Optimized: Use NumPy’s built-in vectorized generator

**Step3: Generate Samples**
- Create n random values
- Scale and shift them using μ and σ

**Step 4: Validate Results**
- Compute mean & standard deviation
- Plot histogram
- Compare with expected values

### Solution Code

````python
# Approach1: Brute Force Approach (Box–Muller Transform)
# Box–Muller converts uniform random variables into normal random variables.
import numpy as np

def generate_normal_bruteforce(mean, std, n):
    u1 = np.random.rand(n)
    u2 = np.random.rand(n)
    
    z = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    return mean + std * z
````

### Alternative Solution

````python
# Approach2: Optimized Approach (Numpy Vectorized)
def generate_normal_optimized(mean, std, n):
    return np.random.normal(loc=mean, scale=std, size=n)
````

### Alternative Approaches

```python
| Approach              | Description                  |
| --------------------- | ---------------------------- |
| Box–Muller            | Mathematical transformation  |
| Ziggurat algorithm    | Faster, low-level method     |
| Central Limit Theorem | Sum of uniform variables     |
| SciPy stats.norm      | Theoretical & sampling tools |
```

### Test Case

````python
# Test Case1: Basic Normal Distribution
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

data = generate_normal_optimized(mean=50, std=10, n=10000)

plt.hist(data, bins=50, density=True)
plt.title("Normal Distribution (μ=50, σ=10)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
````

````python
# Test Case2: Mean & Standard Deviation Validation
print("Sample Mean:", np.mean(data))
print("Sample Std Dev:", np.std(data))
````

````python
# Test Case3: Compare Brute Force vs Optimized
data_bf = generate_normal_bruteforce(0, 1, 10000)
data_opt = generate_normal_optimized(0, 1, 10000)

print("Brute Force Mean:", np.mean(data_bf))
print("Optimized Mean:", np.mean(data_opt))
````

````python
# Test Case4: Small Sample Size (Edge Case)
small_data = generate_normal_optimized(0, 1, 5)
print(small_data)
````

### Expected Outputs
- Generated data follows a normal distribution
- Histogram visually matches bell curve
- Mean and standard deviation align with input values
- Optimized version is faster and preferred
